```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df=pd.read_csv('ds_salaries.csv')
```

```python
print("Duplicate rows:")
print(df.duplicated())

df = df.drop_duplicates()  # Remove duplicate rows
```

```python
df.describe()
```

```python
df.dtypes
```

```python
salary_counts = df['salary'].value_counts().to_frame()
```

```python
salary_counts
```

```python
df['salary_in_usd'].hist(bins=30, edgecolor='black')
plt.gca().set_facecolor('lightblue')
plt.title("Histogram of Data Scientist Salaries")
plt.xlabel("Salary in USD")
plt.ylabel("Frequency")
plt.gca().set_facecolor('lightblue')
plt.show()
```

```python
sns.histplot(data=df, x="salary_in_usd", bins=30, color='green')
plt.title("Histogram of Data Scientist Salaries")
plt.xlabel("Salary in USD")
plt.ylabel("Frequency")
plt.show()
```

```python
# Create the histogram in plotly

fig = px.histogram(df, x="salary_in_usd", nbins=30)
fig.update_layout(
    title='Histogram of Data Scientist Salaries',
    xaxis_title='Salary in USD',
    yaxis_title='Frequency'
)
fig.update_layout(plot_bgcolor='lightblue')
fig.update_traces(marker_color='brown')

fig.show()
```
